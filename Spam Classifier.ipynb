```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r"(?u)\b\w+\b",analyzer = "word", tokenizer = None, preprocessor = None, stop_words = None)
train_spam = vectorizer.fit_transform(train[train['label'] == 1]['v2'])
total_count_spam = train_spam.sum()
spam_feature = vectorizer.get_feature_names()
train_ham = vectorizer.fit_transform(train[train['label'] == 0]['v2']) 
total_count_ham = train_ham.sum()
ham_feature = vectorizer.get_feature_names()
```

```python
def bowFormat(message):
    vectorizer = CountVectorizer(token_pattern=r"(?u)\b\w+\b",analyzer = "word", tokenizer = None, preprocessor = None, stop_words = None)
    words = vectorizer.fit_transform(message)
    word_list = vectorizer.get_feature_names()
    return (words, word_list)
```

```python
def indexFind(a_list, value):
    try:
        return a_list.index(value)
    except :
        return -1
```

```python
def classify(message):
    prob_spam = 1.0
    prob_ham = 1.0
    words, word_list = bowFormat(message)
    if len(word_list) == 0:
        return 1
    for w in word_list:
        index = indexFind(train_spam, w)
        if index is not -1:
            temp = (train_spam[index] + 1)/float(total_count_spam + len(train_spam))
            prob_spam *= temp
        index = indexFind(train_ham, w)
        if index is not -1:
            temp = (train_ham[index] + 1)/float(total_count_ham + len(train_ham))
            prob_ham *= temp
    prob_spam *= len(train[train['label'] == 1])/float(len(train))
    prob_ham *= len(train[train['label'] == 0])/float(len(train))
    return 1 if prob_spam > prob_ham else 0    
```

```python
def calculate ( pred, actual):
    tp = 0
    tn = 0
    fp = 0
    fn = 0
    for i in range(len(pred)):
        if pred[i] == 1 and actual[i] == 1:
            tp += 1
        elif pred[i] == 1 and actual[i] == 0:
            fp += 1
        elif pred[i] == 0 and actual[i] == 1:
            fn += 1
        else:
            tn += 1       
    precision = (tp+1)/float(tp + fp+1)
    recall = (tp+1)/float(tp + fn+1)
    accuracy = (tp+tn+1)/float(len(pred)+1)
    return ( precision, recall, accuracy)
```

```python
import re
def removestop(text):
    pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*')
    text = pattern.sub('', text)
    return text
```

```python
pred = []
for x in range(len(test)):
    message = test.iloc[x]['v2']
    message = removestop(message)
    if len(message) > 10:
        pred.append(classify(message))
    else:
        pred.append(1)
```
